### 4.с помощью данных функций используя функции из модуля itertools сгенерировать 7 фигур, включающих как прямоугольники, так и треугольники и шестиугольники, визуализировать результат.

```python
lst = list(itr.zip_longest(gen_rectangle(3),gen_triangle(3), gen_hexagon(3)))
new_value = gen_rectangle(4)[3]
for i in range(4):
    new_value[i][0] = new_value[i][0] + 2

lst.insert(3, [new_value])

n = 0 
result = []
for i in lst:
    if i == 3:
        result += i
    else:
        result += [i[n%3]]
    n += 1

visualization(-10, 10, result)
```
