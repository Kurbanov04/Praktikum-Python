1) Прямоугольники

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon as pltPolygon
from matplotlib.ticker import MultipleLocator

def gen_rectangle(polygons):
    fig,ax = plt.subplots()
    for polygon in polygons:
        plt_polygon = pltPolygon(polygon, True, facecolor = "blue")
        ax.add_patch(plt_polygon)
       
        
    
    ax.axis('equal')
    ax.set_xlim(0,5) 
    ax.set_ylim(0,3)
    ax.xaxis.set_major_locator(MultipleLocator(base = 0.7))
    plt.axhline(0, color = "black")
    plt.axvline(0.7, color = "black")
    ax.grid()
    
    plt.show()
    
polygons = [((0.7,0), (0.7,0.5), (1.4,0.5), (1.4,0)),
            ((1.4,0), (1.4, 0.5), (2.1, 0.5), (2.1, 0)),
            ((2.1,0), (2.1,0.5), (2.8,0.5), (2.8,0)),
            ((2.8,0),(2.8,0.5),(3.5,0.5),(3.5,0)),
            ((3.5,0),(3.5,0.5),(4.2,0.5),(4.2,0))]

gen_rectangle(polygons)
```

2) Треугольники

```python
def gen_triangle(polygons):
    fig, ax = plt.subplots()
    for polygon in polygons:
        # Создаем треугольный многоугольник
        plt_polygon = pltPolygon(polygon, True, facecolor = "blue")
        ax.add_patch(plt_polygon)
        
    
    ax.axis('equal')
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 3)
    ax.xaxis.set_major_locator(MultipleLocator(base=0.5))
    plt.axhline(0, color="black")
    plt.axvline(0.5, color="black")
    ax.grid()
    plt.show()

# Определяем треугольники, используя три координаты каждый
triangles = [((0.5, 0), (0.75, 0.5), (1, 0)),
             ((1, 0), (1.25, 0.5), (1.5, 0)),
             ((1.5,0),(1.75,0.5),(2,0)),
             ((2,0),(2.25,0.5),(2.5,0)),
             ((2.5,0),(2.75,0.5),(3,0)),
             ((3,0),(3.25,0.5),(3.5,0)),
             ((3.5,0),(3.75,0.5),(4,0)),
             ((4,0),(4.25,0.5),(4.5,0))]

# Вызов функции с треугольниками
gen_triangle(triangles)
```

3) Шестиугольники

```python
from matplotlib.patches import RegularPolygon


def gen_hexagons(polygons):
    fig, ax = plt.subplots()
    
    for polygon in polygons:
        # Extract the six coordinates for each hexagon
        x_coords = [coord[0] for coord in polygon]
        y_coords = [coord[1] for coord in polygon]
        
        # Create the hexagon polygon
        plt_polygon = RegularPolygon((x_coords[0], y_coords[0]), numVertices=6, radius=0.25, edgecolor='black', facecolor='blue')
        ax.add_patch(plt_polygon)
        
    ax.axis('equal')
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 3)
    ax.xaxis.set_major_locator(MultipleLocator(base=0.5))
    plt.axhline(0, color="black")
    plt.axvline(0.35, color="black")
    plt.show()

# Define the hexagons using six coordinates each
hexagons = [((0.625, 0.25), (0.875, 0.5), (0.625, 0.75), (0.375, 0.75), (0.125, 0.5), (0.375, 0.25)),
            ((1.125, 0.25), (1.375, 0.5), (1.125, 0.75), (0.875, 0.75), (0.625, 0.5), (0.875, 0.25)),
            ((1.625, 0.25), (1.875, 0.5), (1.625, 0.75), (1.375, 0.75), (1.125, 0.5), (1.375, 0.25)),
            ((2.125, 0.25), (2.375, 0.5), (2.125, 0.75), (1.875, 0.75), (1.625, 0.5), (1.875, 0.25)),
            ((2.625, 0.25), (2.875, 0.5), (2.625, 0.75), (2.375, 0.75), (2.125, 0.5), (2.375, 0.25)),
            ((3.125, 0.25), (3.375, 0.5), (3.125, 0.75), (2.875, 0.75), (2.625, 0.5), (2.875, 0.25)),
            ((3.625, 0.25), (3.875, 0.5), (3.625, 0.75), (3.375, 0.75), (3.125, 0.5), (3.375, 0.25)),
            ((4.125, 0.25), (4.375, 0.5), (4.125, 0.75), (3.875, 0.75), (3.625, 0.5), (3.875, 0.25))]

# Call the function with the hexagons
gen_hexagons(hexagons)
```

4. С помощью данных функций используя функции из модуля itertools сгенерировать 7 фигур, включающих как прямоугольники, так и треугольники и шестиугольники, визуализировать результат.

```python
import itertools
from matplotlib.patches import Polygon as pltPolygon, RegularPolygon


# Define the polygons with different locations
rectangles = [((0.2,0.2), (0.2,0.7), (0.9,0.7), (0.9,0.2)),
              ((1.3,0.8), (1.3, 1.3), (2, 1.3), (2, 0.8)),
              ((2.6,1.4), (2.6,1.9), (3.3,1.9), (3.3,1.4)),
              ((3.8,0.5),(3.8,1),(4.5,1),(4.5,0.5)),
              ((4.7,0),(4.7,0.5),(5.4,0.5),(5.4,0))]

triangles = [((0, 1), (0.25, 1.5), (0.5, 1)),
             ((1.1, 2), (1.35, 2.5), (1.6, 2)),
             ((2.2, 1), (2.45, 1.5), (2.7, 1)),
             ((3.3, 2), (3.55, 2.5), (3.8, 2)),
             ((4.4, 1), (4.65, 1.5), (4.9, 1)),
             ((0.5, 0), (0.75, 0.5), (1, 0)),
             ((1.5,0),(1.75,0.5),(2,0)),
             ((2.5,0),(2.75,0.5),(3,0)),
             ((3.5,0),(3.75,0.5),(4,0)),
             ((4.5,0),(4.75,0.5),(5,0))]

hexagons = [((0.325, 1), (0.575, 1.25), (0.325, 1.5), (0.075, 1.5), (-0.175, 1.25), (0.075, 1)),
            ((1.025, 2), (1.275, 2.25), (1.025, 2.5), (0.775, 2.5), (0.525, 2.25), (0.775, 2)),
            ((1.825, 1), (2.075, 1.25), (1.825, 1.5), (1.575, 1.5), (1.325, 1.25), (1.575, 1)),
            ((2.625, 2), (2.875, 2.25), (2.625, 2.5), (2.375, 2.5), (2.125, 2.25), (2.375, 2)),
            ((3.425, 1), (3.675, 1.25), (3.425, 1.5), (3.175, 1.5), (2.925, 1.25), (3.175, 1)),
            ((4.225, 2), (4.475, 2.25), (4.225, 2.5), (3.975, 2.5), (3.725, 2.25), (3.975, 2)),
            ((0.875, 0.5), (1.125, 0.75), (0.875, 1), (0.625, 1), (0.375, 0.75), (0.625, 0.5)),
            ((1.875, 0.5), (2.125, 0.75), (1.875, 1), (1.625, 1), (1.375, 0.75), (1.625, 0.5)),
            ((2.875, 0.5), (3.125, 0.75), (2.875, 1), (2.625, 1), (2.375, 0.75), (2.625, 0.5)),
            ((3.875, 0.5), (4.125, 0.75), (3.875, 1), (3.625, 1), (3.375, 0.75), (3.625, 0.5)),
            ((4.875, 0.5), (5.125, 0.75), (4.875, 1), (4.625, 1), (4.375, 0.75), (4.625, 0.5))]

# Define the colors for each polygon type
rect_color = 'red'
tri_color = 'green'
hex_color = 'blue'

# Combine the polygons using itertools
polygons = list(itertools.chain(rectangles, triangles, hexagons))

# Define the function to visualize the polygons
def combine_polygons(polygons):
    fig, ax = plt.subplots()
    for polygon in polygons:
        # Extract the coordinates for each polygon
        x_coords = [coord[0] for coord in polygon]
        y_coords = [coord[1] for coord in polygon]
        
        # Create the appropriate polygon object based on the number of vertices
        if len(polygon) == 4:
            plt_polygon = pltPolygon(polygon, True, edgecolor='black', facecolor=rect_color)
        elif len(polygon) == 3:
            plt_polygon = pltPolygon(polygon, True, edgecolor='black', facecolor=tri_color)
        elif len(polygon) == 6:
            plt_polygon = RegularPolygon((x_coords[0], y_coords[0]), numVertices=6, radius=0.25, edgecolor='black', facecolor=hex_color)
        
        ax.add_patch(plt_polygon)
    
    ax.axis('equal')
    ax.set_xlim(0, 5.5)
    ax.set_ylim(0, 3)
    ax.xaxis.set_major_locator(MultipleLocator(base=0.5))
    #plt.axhline(0, color="black")
    #plt.axvline(0.2, color="black")
    ax.grid()
    plt.show()

# Call the function with the combined polygons
combine_polygons(polygons)
```
